## Homework2. Principal Component Analysis (PCA)

### Goal

Represent digits '3' in 2D

### Requirement

1. convert data from the UCI Optical Recognition of Handwritten Digits Data Set

2. perform PCA over all digit '3' with 2 components

3. plot the PCA results as below (also in page #12 of PCA)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['image.cmap'] = 'Greys'


class DigitDataset(object):
    """ A Simple wrapper class loading UCI dataset.
    
    Parameter:
    ----------
    1. path: [string], path-to-dataset.
    2. n_class: [Scalar], choose which number.
    
    Returns:
    ----------
    1. data: [ndarray]
    2. label: [ndarray]
    """
    
    def __init__(self, path, n_class=None):
        with open(path) as f:
            self.raw = f.readlines()
        self.magic = 21
        self.n_class = n_class
        self.load_datasets()
    
    def _convert(self, str_line):
        str_line = str_line.strip()
        return np.array(list(str_line), dtype=np.uint8)
    
    def _load_samples(self, idx):
        data = np.array([self._convert(i) for i in self.raw[idx: idx+32]])
        label = int(self.raw[idx+32].strip())
        
        return data, label
        
    def load_datasets(self):
        self.data = []
        self.target = []
        
        for i in range(self.magic, len(self.raw), 33):
            data, label = self._load_samples(i)
            self.data.append(data)
            self.target.append(label)
        
        self.data = np.array(self.data)
        self.target = np.array(self.target)
        # In case, we want to access specific number of opt-digit
        if self.n_class:
            idx = np.where(self.target == self.n_class)
            self.data = self.data[idx]
            self.target = self.target[idx]
```

```python
def show_digits(data, n_width=5, n_height=5):
    _, m, n = data.shape
    picture = np.zeros((n_width*m, n_height*n))
    
    for i in range(n_width):
        for j in range(n_height):
            picture[32*i:32*(i+1), 32*j:32*(j+1)] = data[5*j + i]
    
    ax = plt.gca()
    ax.set_xticks(np.arange(0, 32*n_width+1, 32) )
    ax.set_yticks(np.arange(0, 32*n_height+1, 32) )
    ax.set_xticklabels('')
    ax.set_yticklabels('')
    plt.grid(color='r', linestyle='solid', linewidth=2)
    plt.imshow(picture)
```

```python
def PCA(data):
    """
    # Args:
        1. An ndarray(m-by-n) with m samples and n features.
    
    # Returns:
        2. An ndarray(m-by-2), we want to convert it to 2D to visual it.
    """
    # Mean substraction.
    _data = data - np.mean(data, axis=0)
    # Covariance Matrix.
    cov  = np.cov(_data, rowvar = 0)
    # Eigen vector decomposition.
    eigVal, eigVec = np.linalg.eig(cov)
    # Return 2D data
    return np.dot(data, eigVec[:,:2].real),  eigVec[:,:2].real
```

```python
def _nearest_point(data, x, y):
    """data is m-by-n matrix, we want to find nearest point idx to (x,y)
    """
    dist = np.sum((data - np.array([x,y]))**2, axis=1)
    return np.argmin(dist)
```

```python
def show_points(data):
    plt.scatter(data[:, 0], data[:, 1], color='g')
    # find nearest to grid point and plot it.
    for i in range(-4, 6, 2):
        for j in range(-4, 6, 2):
            plt.scatter(data[_nearest_point(data, i, j), 0],
                        data[_nearest_point(data, i, j), 1],
                        color = 'w',
                        edgecolors = 'r',
                        linewidth = 2)
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.title('Principal Component Ananlysis(PCA) in 2D')
    plt.grid()
```

```python
# Loading all 4 subdataset of UCI optic handwritten. 
# Note that each of instance has shape (32, 32)

# training dataset.
dataset1 = DigitDataset('data/optdigits-orig.tra/optdigits-orig.tra', 3)

# cross validation dataset.
dataset2 = DigitDataset('data/optdigits-orig.cv/optdigits-orig.cv', 3)

# Writer independant test dataset.
dataset3 = DigitDataset('data/optdigits-orig.windep/optdigits-orig.windep', 3)

# Writer dependant test dataset.
dataset4 = DigitDataset('data/optdigits-orig.wdep/optdigits-orig.wdep', 3)

data = np.concatenate([dataset1.data, dataset2.data, dataset3.data, dataset4.data])
data = data.reshape((data.shape[0], -1))
```

```python
# Plot data
show_digits(dataset1.data)
```

```python
point, base = PCA(data)
show_points(point)
```

```python
# Plot base1 and base2.
fig = plt.figure()
ax1 = plt.subplot(131)
ax2 = plt.subplot(132)
ax3 = plt.subplot(133)
ax1.imshow(np.mean(data, axis=0).reshape(32, 32))
ax2.imshow(base[:, 0].reshape(32, 32))
ax3.imshow(base[:, 1].reshape(32, 32))

ax1.set_title('mean value')
ax2.set_title('base1 value')
ax3.set_title('base2 value')
plt.show()
```
